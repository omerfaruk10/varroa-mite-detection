```python
model = RTDETR("rtdetr-l.pt")

# Display model information (optional)
model.info()

# Train the model on the COCO8 example dataset for 100 epochs
results = model.train(
    data="/content/Varroa-Mites-Detector-2/data.yaml",
    epochs=100,
    imgsz=640,
    lr0=0.001,          # Öğrenme oranını düşürdük (Varsayılan genelde 0.01)
    amp=False,           # NaN hatasını önlemek için karma hassasiyeti kapattık
    deterministic=False, # Loglardaki o hatayı susturur ve eğitimi rahatlatır
    optimizer='AdamW'    # RT-DETR/Transformer modelleri için genelde daha stabildir
)

```

```python
model = RTDETR("/content/drive/MyDrive/rt-detr_model.pt")
```

```python
metrics = model.val(data='/content/Varroa-1/data.yaml', split='test')

# Sonuçlara erişim
print(metrics.box.map)    # mAP50-95
print(metrics.box.map50)  # mAP50
```
